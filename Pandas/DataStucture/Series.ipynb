###Series


```python
import pandas as pd

list = [1,2,3,4,5,6,7]

s1 = pd.Series(list)

print(s1) 
```

### we can use index of our own and can change the data type also


```python
from pandas.core.interchange import column
list = [1,2,3,4]

s2 = pd.Series(list,index=['a','b','c','s'])

print(s2)
```
