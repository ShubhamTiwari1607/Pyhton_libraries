```python
import numpy as np
import pandas as pd
```

```python
# Generate the resturant ids 

ids = np.arange(20001)
print(ids)
```

```python
q1 = np.random.randint(100000, 300000, 20001)
q2 = q1 + np.random.randint(-20000, 50000, 20001)
q3 = q2 + np.random.randint(-15000, 60000, 20001)
q4 = q3 + np.random.randint(-10000, 70000, 20001)
```

```python
df = pd.DataFrame({
    "ids": ids ,
    "q1":q1 ,
    "q2":q2 ,
    "q3":q3 ,
    "q4": q4
   }
)
```

```python
df.to_csv("sales_data.csv", index=False)
```
