```python
import numpy as np
import pandas as pd 
```

# importing data

```python
Sales_data = np.loadtxt("sales_data.csv", delimiter=",", skiprows=1,dtype=int)
print(Sales_data)
```

```python
print(Sales_data[:2])
```

# Sum of the data


```python
# axis = 0  sum top to bottom in a column
print(np.sum(Sales_data,axis=0))  
#  axis = 1 sum left to right in a row 
print(np.sum(Sales_data,axis=1))
```

```python
# more operation using sum  and slicing

print(np.sum(Sales_data[0:101,1:],axis=0)) 
# above sum the columns top to bottom starting from row 0 to 100 and columns from 1 to last column 
```

# Minimum and Maximum of the data

```python
# minimum sales amongst the restraunt (all rows and all colms except the index) quarter-wise

minSaleAmongstRestraunt = np.min(Sales_data[:,1:],axis=0)

print(minSaleAmongstRestraunt)
```

```python
# minimum sales in the quarters in restraunt
```

```python
minSaleAmongstRestraunt = np.min(Sales_data[:,1:],axis=1)

print(minSaleAmongstRestraunt)
```
